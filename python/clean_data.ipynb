```python
from pathlib import Path
import pandas as pd
import os
```

```python
data_path = Path(os.getcwd()).parent
raw_df = pd.read_csv(f"{data_path}/data/raw_expenses.csv")
df = raw_df.copy()
df.head()
```

```python
print(df.shape)
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
expense_cols = ["Rent","Loan_Repayment","Insurance","Groceries","Transport","Eating_Out","Entertainment","Utilities","Healthcare","Education","Miscellaneous"]
df["total_expenses"] = df[expense_cols].sum(axis = 1)
savings_cols = ["Potential_Savings_Groceries","Potential_Savings_Transport","Potential_Savings_Eating_Out","Potential_Savings_Entertainment","Potential_Savings_Utilities","Potential_Savings_Healthcare","Potential_Savings_Education","Potential_Savings_Miscellaneous"]
df["total_potential_savings"] = df[savings_cols].sum(axis = 1)
df["actual_savings"] = df["Income"] - df["total_expenses"]
df["savings_rate"] = df["actual_savings"] / df["Income"]
df.head()
```

```python
df[df.select_dtypes(include="float").columns] = df.select_dtypes(include="float").round(2)
df.head()
```

```python
df.to_csv(f"{data_path}/data/cleaned_expenses.csv", index=False)
```
